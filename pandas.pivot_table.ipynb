<h1>pandas.pivot_table</h1>

```python
import pandas as pd
import numpy as np

path = '/Users/Sammy/Desktop/'
file = 'MacBookPro.csv'

df = pd.read_csv(path+file, sep=',', thousands=',')  #thousands=',' 옵션으로 천 단위 자리수 구분 콤마 없애고 불러오기
df
```

```python
rst = pd.pivot_table(df, index = ['Processor','Memory']
                       , columns = 'Storage', values = 'Price', aggfunc = np.sum)

# values = ['Price'] 리스트형태로 지정해주면 reindex 컬럼 순서를 재지정할 때 에러가 발생한다

rst
```

```python
# 컬럼 순서 변경
columnsTitles = ['512GB','1TB','2TB','4TB','8TB']
rst = rst.reindex(columns=columnsTitles)
rst
```

```python
rst = pd.pivot_table(df, index = ['Memory','Storage'], values = ['Sale','Amount']
                     , aggfunc = {'Sale': np.sum, 'Amount': np.sum}
                     , margins = True
                    )
rst
# 컬럼 순서 변경
columnsTitles = ['Sale', 'Amount']
rst = rst.reindex(columns=columnsTitles)
rst
```

```python
rst = pd.pivot_table(df, index = ['Memory','Storage'], values = ['Sale','Amount']
                     , aggfunc = {'Sale': np.sum, 'Amount': [min, max, np.sum, np.mean, np.std]})
rst
```
